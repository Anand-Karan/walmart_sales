```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
df = pd.read_csv("walmart_cleaned.csv")
df = df.iloc[:,1:]
df.head()
```

```python
df.Store.unique()
```

```python
df2 = (df.groupby(["Store","Date"])
          .agg({"IsHoliday":"min",
                "Weekly_Sales":"sum",
                "Temperature":"min",
                "Fuel_Price":"min",
                "MarkDown1":"max",
                "MarkDown2":"max",
                "MarkDown3":"max",
                "MarkDown4":"max",
                "CPI":"min",
                "Unemployment":"min",
                "Type":"min",
                "Size":"min"}).reset_index())
```
